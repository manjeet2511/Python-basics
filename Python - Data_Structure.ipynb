# Q3) What are the main differences between lists and tuples in Python?
**Ans:-**The main differences between lists and tuples in Python are:

**Mutability:**Lists are mutable and Tuples: are immutable

**Syntax:**Lists are defined using square brackets '[ ]' and tuples are defined using parentheses '( )'.

**Methods:**Lists have a broader range of built-in methods for modification (e.g., append(), remove(), sort()) whereas tuples have fewer methods, primarily count() and index(), due to their immutability.

**Use Cases:**Lists are suitable for collections where elements are expected to change frequently, such as storing user inputs or dynamic data and tuples are ideal for representing fixed collections of related items, like coordinates, database records, or function arguments that should not be altered.

# Q8) What is a hash table, and how does it relate to dictionaries in Python?
**Ans:-**A hash table is a data structure that implements an associative array, also known as a dictionary or map. It stores key-value pairs and uses a hash function to map keys to indices in an internal array, allowing for efficient data retrieval.
Python's built-in dict type is an implementation of a hash table. When you create a dictionary in Python and add key-value pairs, Python internally uses a hash table to manage this data.

# Q14) How does the “in” keyword work differently for lists and dictionaries?
**Ans:-** The in keyword in Python is used for membership testing, but its behavior differs for lists and dictionaries due to their underlying data structures and how they store elements.

**For list:-**
When in is used with a list, it checks if a given element exists as one of the list's values.

**For Dictionaries:-**
When in is used with a dictionary, it checks if a given item exists as one of the dictionary's keys. It does not directly check for values.

# Q15) Can you modify the elements of a tuple? Explain why or why not.
**Ans:-** No we can not modify the elements of a tuple because tuples are immutable, meaning once they are created, their contents—such as items, order, and size—cannot be changed. Tuples are intended to represent fixed collections.

# Q16) What is a nested dictionary, and give an example of its use case.
**Ans:-** A nested dictionary in Python is simply a dictionary whose values include other dictionaries. This structure allows you to represent hierarchical or multi-level data where each key maps to a more detailed grouping.

Example of its use case:
Representing data for a system that manages information about multiple students, where each student has various attributes including personal details and academic performance. Here student data is nested dictionary.

#Q17) Describe the time complexity of accessing elements in a dictionary.
**Ans:-**Accessing elements in a dictionary, which is typically implemented as a hash table, generally exhibits a time complexity of O(1) in the average case. This means the time taken to retrieve an element by its key is constant and does not significantly increase with the size of the dictionary.

# Q18) In what situations are lists preferred over dictionaries?
**Ans:-** we prefer using a list over a dictionary in situations where the order of items matters, and we need to process elements by their position or sequence for instance, iterating in a specific order, slicing, or sorting values. Lists support duplicate entries, are easy to modify with methods like .append() or .insert(), and offer constant time access by index.


#Q19) Why are dictionaries considered unordered, and how does that affect data retrieval?
**Ans:-** Dictionaries are considered unordered because they do not guarantee that the order in which items are inserted will be the same order in which they are accessed or retrieved. This lack of inherent order affects data retrieval because you cannot rely on accessing elements by their position as you would do with lists or tuples.
